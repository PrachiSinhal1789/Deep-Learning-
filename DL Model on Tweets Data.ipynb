```python
import pandas as pd
import numpy as np
import regex as re
import nltk   #natural language toolkit
nltk.download('stopwords')
from nltk.corpus import stopwords  #corpus means collection
stop=stopwords.words('english')
```

```python
df=pd.read_csv('Tweets.csv')
```

```python
df.head()
```

```python
df=df[['text','airline_sentiment']]
```

```python
df.head()
```

## Data Cleaning

```python
def clean_data(text):
    sentence=text.lower()   # converts the entire text into lower case
    sentence=re.sub('[^a-z]',' ',sentence)   # substitues all non alphabets symbol like @ with space
    words=sentence.split()  # splits the unique words in a sentence and stores in a list
    imp_words=[w for w in words if w not in stop]  #this filters out all the words which are not stop words
    return ' '.join(imp_words)    #join all words with space to form a sentence
```

```python
df['cleaned_text']=df['text'].apply(lambda x: clean_data(x))  #we apply lambda so that this applies on each row
```

```python
df.head()
```

```python
df=df[['cleaned_text','airline_sentiment']]
```

```python
df.to_csv('Cleaned_TweetsData.csv')
```
