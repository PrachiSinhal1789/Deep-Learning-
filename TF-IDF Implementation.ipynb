```python
import pandas as pd
import numpy as np
```

```python
docs=['I am learning Deep Learning',
     'It is fun to Code',
     'Sunrises are Beautiful',
     'but not more Beautiful than my sleep',
     'We must accept ourselves who we are, until we are serial killers',
     'No one wants to come back to India unless he is aayush',
     'All swipe and no match let me teach you how to batch']
```

### TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf_vec=TfidfVectorizer()
```

```python
tf_vector=tf_vec.fit_transform(docs)
```

```python
tf_vector.toarray()
```

### TF-IDF Transformer

```python
from sklearn.feature_extraction.text import TfidfTransformer,CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
cv_vect=cv.fit_transform(docs)
```

```python
cv_vect.toarray()
```

```python
tf_transform=TfidfTransformer()
```

```python
tf_trans_vect=tf_transform.fit_transform(cv_vect)
```

```python
tf_trans_vect.toarray()
```
